# üìù Zepto Dataset - Exploratory Data Analysis (EDA)

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("zepto.csv")
df.head()
```

## 1. Basic EDA

```python
print("\n--- Basic Info ---")
df.info()
```

```python
print("\n--- Null Values ---")
df.isnull().sum()
```

```python
print("\n--- Summary Statistics ---")
df.describe(include='all')
```

## 2. NumPy Questions

```python
ratings = df['Rating'].values
delivery_times = df['Delivery Time (min)'].values

print("Mean Rating =", np.mean(ratings))
```

```python
print("Median Delivery Time =", np.median(delivery_times))
```

```python
print("Std Dev of Ratings =", np.std(ratings))
```

```python
print("Minimum Rating =", np.min(ratings))
```

```python
print("Unique Delivery Times =", np.unique(delivery_times).size)
```

```python
print("Correlation between Ratings & Delivery Time =", np.corrcoef(ratings, delivery_times)[0,1])
```

## 3. Pandas Questions

```python
print("\n", df.groupby('Order Type')['Rating'].mean())
```

```python
print("\n", df['Customer Feedback Type'].value_counts())
```

```python
print("\n", df.groupby('Location')['Delivery Time (min)'].max())
```

```python
print("\n", df.groupby('Agent Name')['Customer Service Rating'].mean().sort_values(ascending=False).head())
```

```python
print("Avg Delivery Time per Order Type\n", df.groupby('Order Type')['Delivery Time (min)'].mean())
```

```python
print("Most Common Location =", df['Location'].mode()[0])
```

```python
print("Incorrect Orders =", (df['Order Accuracy'].eq('Incorrect').mean()) * 100)
```

```python
print("Orders with Discount & Rating>4\n", df[(df['Discount Applied']=='Yes') & (df['Rating']>4)].head())
```

## 4. Matplotlib Visualizations

```python
plt.hist(ratings, bins=20, edgecolor='black')
plt.title('Histogram of Ratings')
```

```python
df['Order Type'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True)
plt.title('Order Type Distribution')
```

```python
df['Customer Feedback Type'].value_counts().plot(kind='bar')
plt.title('Customer Feedback Type')
```

```python
plt.plot(df['Delivery Time (min)'].head(50))
plt.grid(True)
plt.title('Delivery Times (First 50 Orders)')
```

```python
plt.scatter(df['Delivery Time (min)'], df['Rating'], alpha=0.5)
plt.xlabel('Delivery Time (min)')
plt.ylabel('Rating')
plt.title('Scatter: Delivery Time vs Rating')
```

## 5.Seaborn Visualizations

```python
feedback_price = pd.crosstab(df['Price Range'], df['Customer Feedback Type'])
feedback_price.plot(kind='bar', stacked=True, ax=plt.gca())
plt.title('Stacked Bar: Feedback vs Price Range')
```

```python
sns.scatterplot(x='Delivery Time (min)', y='Rating', data=df)
plt.title('Scatter: Delivery Time vs Rating')
```

```python
sns.lineplot(x=range(50), y=df['Rating'].head(50) ,marker='o', color='blue')
plt.title('Line Plot of Ratings (First 50)')
```

```python
sns.barplot(x='Order Type', y='Rating', data=df)
plt.title('Bar: Order Type vs Rating')
```

```python
sns.boxplot(x='Price Range', y='Rating', data=df)
plt.title('Box: Price Range vs Rating')
```

```python
sns.violinplot(x='Discount Applied', y='Delivery Time (min)', data=df)
plt.title('Violin: Discount vs Delivery Time')
```

```python
sns.countplot(x='Customer Feedback Type', data=df,palette='pastel')
plt.title('Count: Feedback Types')
```

```python
sns.kdeplot(ratings ,  fill=True, color='purple')
plt.grid(True)
plt.title('KDE Plot of Ratings')
```

```python
sns.histplot(ratings, kde=True ,  color='coral')
plt.title('Distribution Plot of Ratings')
```
